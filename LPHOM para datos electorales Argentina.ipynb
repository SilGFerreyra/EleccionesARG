**Descargamos datos**

```r
###Trabajamos con archivo de elecciones generales 2019 para presidente
gen19 <- read.csv ("https://raw.githubusercontent.com/SilGFerreyra/EleccionesARG/main/gen19_presiok.csv")
head (gen19, n=50)
```

```r
####Trabajamos con archivo de elecciones generales 2023 para presidente
gen23 <- read.csv ("https://raw.githubusercontent.com/SilGFerreyra/EleccionesARG/main/gen23_presiok.csv")
head (gen23, n= 50)
```

**Preparamos tablas para analizar transferencia electoral**

**Elecciones 2019**

```r
gen19_presidente_provincia <- gen19 %>%
     mutate(agrupacion_nombre = ifelse(agrupacion_nombre == "", "BLANCO_NULO", agrupacion_nombre))%>%
      mutate(agrupacion_nombre = as.factor(agrupacion_nombre)) %>% #me permite que tome los datos como de esa columna como categorías, no solo como texto
       group_by(distrito_id, agrupacion_nombre) %>%
          summarise(votos = sum (votos_cantidad))

head (gen19_presidente_provincia,  n =25)

###Puedo chequear en resultados.gob.ar
```

```r
gen19_electores<- gen19 %>%
  filter (agrupacion_nombre == "JUNTOS POR EL CAMBIO") %>% ####coloco cualquier agrupación porque es un número que se repite en cada mesa
  group_by(distrito_id) %>%
  summarise(total_electores = sum(mesa_electores))

head (gen19_electores, n=25)
```

```r
gen19_presidente_provincia_full <- full_join(gen19_presidente_provincia, gen19_electores,  by = "distrito_id")

head (gen19_presidente_provincia_full)
```

```r
#Damos vuelta los datos

gen19_presidente_provincia_pivot<- gen19_presidente_provincia_full %>%
  pivot_wider(names_from = agrupacion_nombre, values_from = votos)

head (gen19_presidente_provincia_pivot)
```

```r
gen19_presidente_provincia_pivot <- gen19_presidente_provincia_pivot %>% ungroup() %>%
  mutate(ABSTENCION = (total_electores-rowSums(select(., c("BLANCO_NULO", "CONSENSO FEDERAL", "FRENTE DE IZQUIERDA Y DE TRABAJADORES - UNIDAD", "FRENTE DE TODOS", "FRENTE NOS", "JUNTOS POR EL CAMBIO", "UNITE POR LA LIBERTAD Y LA DIGNIDAD")))))

head (gen19_presidente_provincia_pivot)
```

**Elecciones 2023**

```r
gen23_presidente_provincia <- gen23 %>%
    mutate(agrupacion_nombre = ifelse(agrupacion_nombre == "", "BLANCO_NULO", agrupacion_nombre))%>%
     mutate(agrupacion_nombre = as.factor(agrupacion_nombre)) %>%
      group_by(distrito_id, agrupacion_nombre) %>%
          summarise(votos = sum (votos_cantidad))

head (gen23_presidente_provincia, n=50)
```

```r
gen23_electores <- gen23 %>%
     filter (agrupacion_nombre == "LA LIBERTAD AVANZA") %>% ####coloco cualquier agrupación porque es un número que no varía
       group_by(distrito_id) %>%
        summarise(total_electores = sum(mesa_electores))

head (gen23_electores, n=25)
```

```r
gen23_presidente_provincia_full <- full_join(gen23_presidente_provincia, gen23_electores,  by = "distrito_id")

head (gen23_presidente_provincia_full)
```

```r
#Damos vuelta los datos

gen23_presidente_provincia_pivot<- gen23_presidente_provincia_full %>%
  pivot_wider(names_from = agrupacion_nombre, values_from = votos)

head (gen23_presidente_provincia_pivot)
```

```r
gen23_presidente_provincia_pivot <- gen23_presidente_provincia_pivot %>% ungroup() %>%
  mutate(ABSTENCION = (total_electores-rowSums(select(., c("BLANCO_NULO", "HACEMOS POR NUESTRO PAIS", "FRENTE DE IZQUIERDA Y DE TRABAJADORES - UNIDAD", "JUNTOS POR EL CAMBIO", "LA LIBERTAD AVANZA", "UNION POR LA PATRIA")))))

head (gen23_presidente_provincia_pivot, n=25)
```
